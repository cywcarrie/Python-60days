```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
import pandas as pd
import numpy as np
df = pd.read_csv('boston.csv')
print(df)
df.boxplot()
print('TAX, B')
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
df.plot.scatter(x = 'NOX', y = 'DIS')
print(df[['NOX', 'DIS']].corr(method = 'pearson'))
np.corrcoef(df['NOX'], df['DIS'])
print('負相關')
```
